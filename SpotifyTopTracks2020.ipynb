```python
import json
import time
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import numpy as np
import requests


CLIENT_ID = 'aa41d9f3e6f742b5b04add4a7ad1755c'
CLIENT_SECRET = '3341ddbf1143486b98e7bbc3f9c9a502'

AUTH_URL = 'https://accounts.spotify.com/api/token'

# POST
auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']

headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}

client_credentials_manager = SpotifyClientCredentials(CLIENT_ID, CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
def getTrackIDs(playlist_id):
    ArtistName = []
    ArtistID = []
    ArtistPopularity = []
    ArtistFollowers = []
    TrackTitle = []
    TrackID = []
    TrackPopularity = []
    playlist = sp.playlist(playlist_id)        
    for item in playlist['tracks']['items']:
        track_popularity = item['track']['popularity']
        artist = item['track']['artists'][0]['name'] 
        artist_id = item['track']['artists'][0]['id']
        track = item['track']['name'] 
        track_id = item['track']['id']
        
        artistPop = sp.search(q = artist, type='artist', limit=1)        
        artist_followers = artistPop['artists']['items'][0]['followers']['total'] # followers
        artist_popularity = artistPop['artists']['items'][0]['popularity'] # pop
        
        TrackPopularity.append(track_popularity)
        ArtistPopularity.append(artist_popularity)
        ArtistFollowers.append(artist_followers)
        ArtistName.append(artist)
        ArtistID.append(artist_id)
        TrackTitle.append(track)
        TrackID.append(track_id)

    
    trackData = pd.DataFrame({'Artist Name':ArtistName,
                              'Artist ID':ArtistID,
                              'Artist Popularity': ArtistPopularity,
                              'Artist Followers': ArtistFollowers,
                              'Track Title':TrackTitle,
                              'Track ID':TrackID,
                              'Track Popularity': TrackPopularity
                            })
    trackData.index = np.arange(1, len(trackData) + 1)
    trackData.index.name = 'Rank'

    return trackData
```

```python
def getTrackCharacteristics(trackData):
    for index, row in trackData.iterrows():
        characteristics = requests.get('https://api.spotify.com/v1/audio-features/' + row['Track ID'], headers=headers)
        characteristics = characteristics.json()
        
        trackData.loc[index,'Danceability'] = characteristics.get('danceability')
        trackData.loc[index,'Energy'] = characteristics.get('energy')
        trackData.loc[index,'Key'] = characteristics.get('key')
        trackData.loc[index,'Loudness'] = characteristics.get('loudness')
        trackData.loc[index,'Mode'] = characteristics.get('mode')
        trackData.loc[index,'Speechiness'] = characteristics.get('speechiness')
        trackData.loc[index,'Acousticness'] = characteristics.get('acousticness')
        trackData.loc[index,'Instrumentalness'] = characteristics.get('instrumentalness')
        trackData.loc[index,'Liveness'] = characteristics.get('liveness')
        trackData.loc[index,'Valence'] = characteristics.get('valence')
        trackData.loc[index,'Tempo'] = characteristics.get('tempo')
        trackData.loc[index,'Duration ms'] = characteristics.get('duration_ms')
        trackData.loc[index,'Time Signature'] = characteristics.get('time_signature')  
    
    trackData[['Danceability',
              'Energy',
              'Key',
              'Loudness',
              'Mode',
              'Speechiness',
              'Acousticness',
              'Instrumentalness',
              'Liveness',
              'Valence',
              'Tempo',
              'Duration ms',
              'Time Signature',
              'Artist Popularity',
              'Artist Followers',
              'Track Popularity']] = trackData[['Danceability',
              'Energy',
              'Key',
              'Loudness',
              'Mode',
              'Speechiness',
              'Acousticness',
              'Instrumentalness',
              'Liveness',
              'Valence',
              'Tempo',
              'Duration ms',
              'Time Signature',
              'Artist Popularity',
              'Artist Followers',
              'Track Popularity']].apply(pd.to_numeric)

    trackData.to_csv('spotifyData.csv', sep='\t')
    return trackData
        
```
